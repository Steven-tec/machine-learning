Dada una lista con elementos duplicados, escribir un programa que muestre una nueva lista
con el mismo contenido que la primera pero sin elementos duplicados. Para este ejercicio, no
puedes hacer uso de objetos de tipo ‘Set’.


```python

lista_original = [1, 2, 3, 2, 4, 1, 5, 3, 6, 4, 7, 5]
print("Lista original:")
print(lista_original)

lista_sin_duplicados = []

for elemento in lista_original:
    if elemento not in lista_sin_duplicados:
        lista_sin_duplicados.append(elemento)
        
print("\nLista sin duplicados:")
print(lista_sin_duplicados)
```
