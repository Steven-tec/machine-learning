## 1. Estructuras de control  
### 1.1 Selección (sentencias condicionales)

- Selección de una de varias alternativas en base a alguna condición.  
- Indentación para estructurar el código.  
- Importante el símbolo `:`  


```
# Asigno el valor uno a la variable x
x = int(input("Ingrese el valor de x"))

# Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

- Las condiciones son expresiones que se evalúan a **True** o **False**.  
- Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.


```
x = 3
y = [1, 2, 3]

print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

- Se pueden introducir más ramas en sentencias condicionales a través de la palabra clave **elif**.


```
x = -3

if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

- Anidamiento.


```
val = 120

if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

## 2. Switch


```
a = -6

match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```
a = 'l'  # week day

match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

## Expresión ternaria

- Estructura if-else condensada en una línea.  
- Se recomienda usar solo en casos sencillos.


```
# Ejemplo: cálculo del valor absoluto de un número
x = -3
resultado = x if x >= 0 else -x
print(resultado)
```

```
val = -3

if val >= 0:
    resultado = val
else:
    resultado = -val

print(resultado)
```

## 3. Iteración (bucles)

- Repetición de un bloque de código.  
- Dos tipos: **while** y **for**.


### Bucle while


```
indice = 0
numeros = [9, 4, 7, 1, 2]

while indice < 3:
    print(numeros[indice])
    indice += 1
```

```
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1

print('Contador:', contador)
print('Indice:', indice)
```

```
nombre = 'Pablo'

while nombre:
    print(nombre)
    nombre = nombre[1:]
```

### Bucle for


```
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

for pelicula in peliculas:
    print(pelicula)
```

```
for i in range(10):
    print(i)
```

## 4. List Comprehensions


```
[char*3 for char in "Enrique"]
```

```
[y for y in range(9) if y % 2 == 0]
```

```
[0 if x % 2 == 0 else x for x in range(9)]
```
