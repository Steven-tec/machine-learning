## Página 2

### Comentarios en Python  
Los comentarios son texto ignorado por el intérprete y sirven para documentar el código.

**Tipos de comentarios**
- Comentarios de una línea (`#`)
- Comentarios de varias líneas (triples comillas)


```
# Esto es un comentario de una línea
print('Hello world')
```

## Página 3

Ejemplo de comentario multilínea en una función.

```
def producto(x, y):
    '''
    Esta función recibe dos números como parámetros
    y devuelve su producto.
    '''
    return x * y
```

## Página 4

### Literales, variables y objetos  
Python trabaja con **objetos** y **operaciones** sobre ellos.


```
print(4)
print(4.2)
print('Hello world!')
print(False)
```

## Página 5

Literales compuestos: listas, diccionarios, conjuntos y tuplas.

```
print([1,2,3,3])
print({'Nombre':'John Doe','edad':30})
print({1,2,3})
print((4,5))
```

## Página 6

Variables y asignación en Python.

```
a = 5
print(a)
print(type(a))
```

## Página 7

Tipado dinámico: una variable puede cambiar de tipo.

```
a = 3
print(a, type(a))
a = 'Pablo García'
print(a, type(a))
a = 4.5
print(a, type(a))
```

## Página 8

Identidad de objetos y función id().

```
a = 3
print(id(a))
a = 'Hola'
print(id(a))
```

## Página 9

Tipos de datos básicos: booleanos.

```
a = True
b = False
print(a, type(a))
print(b, type(b))
```

## Página 10

Tipos numéricos en Python.

```
print(2)
print(3.4)
print(2+4j)
print(1/2)
```

## Página 11

Strings: secuencias inmutables.

```
s = 'John "ee" Doe'
print(s[0])
print(s[-1])
print(s[1:8:2])
print(s + 'e')
```

## Página 12

Conversión entre tipos (casting).

```
a = int(2.8)
b = float('3')
c = str(2)
print(a, b, c)
```

## Página 13

Operadores aritméticos.

```
x = 3
y = 2
print(x+y, x-y, x*y, x/y, x//y, x**y)
```

## Página 14

Operadores de comparación.

```
x = 10
y = 12
print(x > y)
print(x == y)
print(x <= y)
```

## Página 15

Operadores lógicos.

```
x = True
y = False
print(x and y)
print(x or y)
print(not x)
```

## Página 16

Listas: creación y operaciones básicas.

```
letras = ['a','b','c']
letras.append('d')
print(letras)
```

## Página 17

Acceso y slicing en listas.

```
letras = ['a','b','c','d','e']
print(letras[1:4])
print(letras[::2])
```

## Página 18

Diccionarios: pares clave-valor.

```
persona = {'Nombre':'Carlos','Edad':34}
print(persona['Nombre'])
```

## Página 19

Modificar y agregar elementos a un diccionario.

```
persona['Ciudad'] = 'Valencia'
persona['Edad'] += 1
print(persona)
```

## Página 20

Conjuntos (sets).

```
set1 = {1,2,3,3,4}
print(set1)
```

## Página 21

Operaciones con conjuntos.

```
a = {1,2,3}
b = {2,3,4}
print(a | b)
print(a & b)
print(a - b)
```

## Página 22

Tuplas: colecciones inmutables.

```
t = (1,2,3)
print(t)
print(t[0])
```

## Página 23

Desempaquetado de tuplas.

```
t = (1,2,3)
a,b,c = t
print(a,b,c)
```

## Página 24

Secuencias: range y enumerate.

```
for i, v in enumerate(['a','b','c']):
    print(i, v)
```

## Página 25

zip para combinar colecciones.

```
nombres = ['Ana','Luis']
edades = [20,30]
for n,e in zip(nombres, edades):
    print(n,e)
```

## Página 26

Mutabilidad vs inmutabilidad.

```
lista = [1,2,3]
tupla = (1,2,3)
lista[0] = 100
print(lista)
print(tupla)
```

## Página 27

Identidad de objetos mutables.

```
a = [1,2,3]
b = a
b.append(4)
print(a)
```

## Página 28

Referencias en objetos inmutables.

```
s = 'Hola'
t = s
s += ' Mundo'
print(s)
print(t)
```
