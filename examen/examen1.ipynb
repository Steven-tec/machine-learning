1. Dada la lista [1,2,3,4,5,6,7,8] escribe un programa que, a partir de esta lista, obtenga la lista
[8,6,4,2] y la muestre por pantalla.

```python

lista = [1, 2, 3, 4, 5, 6, 7, 8]
resultado = lista[::-2]
print(resultado)
```

2. Escribe un programa que, dada una lista de strings L, un string s perteneciente a L y un
string t, reemplace s por t en L. El programa debe mostrar la lista resultante por pantalla.

```python

L = ["perro", "gato", "loro", "pez"]

s = "gato"
t = "conejo"

for i in range(len(L)):
    if L[i] == s:
        L[i] = t

print(L)
```

3. Dado un diccionario cuyos elementos son pares de tipo string y numérico (es decir, las claves
son de tipo ‘str’ y los valores son de tipo ‘int’ o ‘float’), escribe un programa que muestre
por pantalla la clave cuyo valor asociado representa el valor númerico más alto de todo el
diccionario. Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’: -5} la respuesta
sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y -5.

```python

diccionario = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}

clave_maxima = max(diccionario, key=diccionario.get)

print("La clave con el valor más alto es:", clave_maxima)
```

4.Escribe una función que reciba un número entero positivo como parámetro y devuelva una
lista que contenga los 5 primeros múltiplos de dicho número. Por ejemplo, si la función recibe
el número 3, devolverá la lista [3, 6, 9, 12, 15]. Si la función recibe un parámetro incorrecto
(por ejemplo, un múmero menor o igual a cero), mostrará un mensaje de error por pantalla
y devolverá una lista vacía.

```python
def primeros_cinco_multiplos(n):
    if n <= 0:
        print("Error: el número debe ser un entero positivo mayor que cero.")
        return []

    multiplos = []
    for i in range(1, 6):
        multiplos.append(n * i)

    return multiplos


print(primeros_cinco_multiplos(3))
print(primeros_cinco_multiplos(-2))
```
